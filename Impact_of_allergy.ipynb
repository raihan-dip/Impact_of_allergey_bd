```python
from sklearn.preprocessing import LabelEncoder
le_sex=LabelEncoder()
le_living_place=LabelEncoder()
le_affected=LabelEncoder()
le_type=LabelEncoder()
le_med=LabelEncoder()
le_road=LabelEncoder()
```

```python
df2['sex']=le_sex.fit_transform(df2['Sex'])
df2['place']=le_living_place.fit_transform(df2['Living_place'])
df2['affected']=le_affected.fit_transform(df2['Allergy_affected'])
df2['type']=le_type.fit_transform(df2['Allergy_type'])
df2['med']=le_med.fit_transform(df2['Taking_medicine'])
df2['road_construction']=le_road.fit_transform(df2['ongoing_road_construction'])
df2.head()
```

```python
df3=df2.drop(['Sex','Living_place','Allergy_affected','Allergy_type','Taking_medicine','ongoing_road_construction'],axis='columns')
df3.head()
```

```python
df4=df3.drop(['type','med','road_construction'],axis='columns')
df4
```

```python
X=df4.drop(['affected'],axis='columns')
X
```

```python
y=df4.affected
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
```

```python
model1.fit(X_train,y_train)
```

```python
model1.score(X_train,y_train)
```

```python
from sklearn.svm import SVC
model2=SVC()
```

```python
model2.fit(X_train,y_train)
```

```python
model2.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier()
```

```python
model3.fit(X_train,y_train)
```

```python
model3.score(X_test,y_test)
```
